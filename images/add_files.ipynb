# Trade analys
In this notebook file are present code to supervise the weekly reports of the category trade analysis of a shop. the reports are in an excel file where in every sheet the sales of the week are written and divided by category. 

Those reports are written every week, the goal of this project is to find an easy and faster way read, manipulate, and supervise the trade.

3 dataframes will be created. first one with the weekly sales of every single product categorized, Second the total weekly sales of each category. and last the total sale of each category.


the first step is importing the module needed for the analysis, import the xlsx file and write the week.


```python
import pandas as pd
import xlrd
import numpy as np
import matplotlib.pyplot as plt
xls = pd.ExcelFile('Category and Dept Analysis Week 05 2021.xlsx')
week=5
```

# Create a dataframe with all categories 
Steps:

 1) Imprt the tables.
 
 2) Add a colum to each table where indicate the name of the category.
 
 3) Concat all the table to create a dataframe.
 
 4) clean data (delete na rows)
 
 5) Add columns
 
 6) create the tables
 

```python
# 1) and 2)
Cheese = pd.read_excel(xls, 'Cheese', index_col= None, skiprows = 4, nrows=250,usecols='A:R')
Cheese['Group']= "Cheese"
Bread = pd.read_excel(xls, 'Bread', index_col= None, skiprows = 4, nrows=250,usecols='A:R')
Bread['Group']= "Bread"
Dairy_Egg = pd.read_excel(xls, 'Dairy and Egg', index_col= None, skiprows = 4, nrows=250,usecols='A:R')
Dairy_Egg['Group']= "Dairy and Egg"
Meat = pd.read_excel(xls, 'Meat and Fish', index_col= None, skiprows = 5, nrows=250,usecols='A:R')
Meat['Group']= "Meat and Fish"
Black = pd.read_excel(xls, 'Black Barn', index_col= None, skiprows = 5, nrows=250,usecols='A:R')
Black['Group']= "Black Barn"
Fruit = pd.read_excel(xls, 'Fruit and Veg', index_col= None, skiprows = 5, nrows=250,usecols='A:R')
Fruit['Group']= "Fruit and Veg"
Patisserie = pd.read_excel(xls, 'Patisserie', index_col= None, skiprows = 5, nrows=250,usecols='A:R')
Patisserie['Group']= "Patisserie"
Charcuterie = pd.read_excel(xls, 'Charcuterie', index_col= None, skiprows = 4, nrows=250,usecols='A:R')
Charcuterie['Group']= "Charcuterie"
FTG = pd.read_excel(xls, 'FTG', index_col= None, skiprows = 4, nrows=250,usecols='A:R')
FTG['Group']= "FTG"
Juicesy = pd.read_excel(xls, 'Fresh Juices', index_col= None, skiprows = 4, nrows=250,usecols='A:R')
Juicesy['Group']= "Fresh Juices"
Wine = pd.read_excel(xls, 'Wine Spirits', index_col= None, skiprows = 5, nrows=250,usecols='A:R')
Wine['Group']= "Wine Spirits"
Deli = pd.read_excel(xls, 'Deli', index_col= None, skiprows = 5, nrows=250,usecols='A:R')
Deli['Group']= "Deli"
TA_drinks = pd.read_excel(xls, 'Takeaway Drinks', index_col= None, skiprows = 4, nrows=250,usecols='A:R')
TA_drinks['Group']= "TA_drinks"
#-------------------------------------------------------------------------------------------------------------
# 3)

frames = [Cheese, Bread, Dairy_Egg, Meat, Black, Fruit, Patisserie, Charcuterie, FTG, Juicesy, Wine, Deli, TA_drinks]
result = pd.concat(frames)

products= result.dropna(subset = ['SKU', 'Description']) #4)
#5)

products['Initial_MarTY'] = (products['Net £']/products['QTY Sales'])
products['Initial_MarLY'] = (products['Net £.1']/products['QTY Sales.1'])
products['Week']= week
#--------------------------------------------------------------------------------------------------------------------
#6)

sumg = products.groupby(['Group'])['Net £', 'Net £.1', 'Cost £', 'Cost £.1'].sum()
sumg['Wastage%TY'] = (sumg ['Cost £']/(sumg ['Cost £']+ sumg['Net £']) * 100)
sumg['Wastage%LY'] = (sumg ['Cost £.1']/(sumg ['Cost £.1']+ sumg['Net £.1']) * 100)
sumg['Week']= week

#-----------------------------------------------------------------------------------------------------------
tot = sumg.groupby(['Week'])['Net £', 'Net £.1', 'Cost £', 'Cost £.1'].sum()
tot['Wastage%TY'] = (tot ['Cost £']/(tot ['Cost £']+ tot['Net £']) * 100)
tot['Wastage%LY'] = (tot ['Cost £.1']/(tot['Cost £.1']+ tot['Net £.1']) * 100)


```

# Save the tables in csv files
the three new tables create are saved in three different csv files for future analysis.

```python
file = open("products_list2021.csv", "a")
products.to_csv('products_list2021.csv', mode='a', header=False)
file.close()
```

```python
file2 = open("sumg.csv", "a")
sumg.to_csv('sumg.csv', mode='a', header=False)
file2.close()
```

```python
file3 = open("tot.csv", "a")
tot.to_csv('tot.csv', mode='a', header=False)
file3.close()
```
